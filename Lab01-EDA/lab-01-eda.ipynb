## 1.1 What is Exploratory Data Analysis (EDA)?

**EDA** is the process of investigating and understanding your data before building any models. Think of it as getting to know your data - like meeting someone new and learning about their background, interests, and characteristics.

### Why is EDA Important for Security?

In security analysis, EDA helps you:
- **Understand attack patterns** - What types of attacks are most common?
- **Identify data quality issues** - Are there missing values or errors?
- **Discover insights** - What can the data tell us about security threats?
- **Guide model building** - What features should we focus on?

### EDA Process Overview

```
1. Load Data ‚Üí 2. Basic Info ‚Üí 3. Data Quality ‚Üí 4. Visualizations ‚Üí 5. Insights
```

## 1.2 Security Data Types

Security data comes in many different forms. Let's explore the main types you'll encounter:

### 1. Network Traffic Data
- **What it is:** Information about network communications
- **Examples:** IP addresses, ports, protocols, packet sizes
- **Use cases:** Intrusion detection, DDoS detection, network monitoring

### 2. Log Data
- **What it is:** Records of system events and activities
- **Examples:** System logs, application logs, security logs
- **Use cases:** Incident response, threat hunting, compliance

### 3. File Data
- **What it is:** Information about files and their characteristics
- **Examples:** File size, entropy, API calls, strings
- **Use cases:** Malware detection, file classification

### 4. User Behavior Data
- **What it is:** Information about how users interact with systems
- **Examples:** Login times, access patterns, resource usage
- **Use cases:** Insider threat detection, fraud detection

### 5. Email Data
- **What it is:** Information about email communications
- **Examples:** Sender, subject, content, attachments
- **Use cases:** Phishing detection, spam filtering

### 6. Threat Intelligence Data
- **What it is:** Information about known threats and indicators
- **Examples:** IP addresses, domains, file hashes
- **Use cases:** Threat hunting, incident response

## 1.4 Key EDA Techniques

Now let's learn the main techniques you'll use to explore your data:

### 1. **Data Overview**
- **What it is:** Getting basic information about your dataset
- **What you learn:** Size, structure, data types, memory usage
- **How to do it:** `df.info()`, `df.shape`, `df.head()`

### 2. **Missing Values Analysis**
- **What it is:** Finding and understanding missing data
- **What you learn:** Data completeness, potential issues
- **How to do it:** `df.isnull().sum()`, `df.isnull().mean()`

### 3. **Statistical Summary**
- **What it is:** Basic statistics for numerical data
- **What you learn:** Central tendency, spread, distribution
- **How to do it:** `df.describe()`, `df.mean()`, `df.std()`

### 4. **Distribution Analysis**
- **What it is:** Understanding how data is distributed
- **What you learn:** Data patterns, outliers, normality
- **How to do it:** Histograms, box plots, density plots

### 5. **Correlation Analysis**
- **What it is:** Finding relationships between variables
- **What you learn:** Which features are related, multicollinearity
- **How to do it:** Correlation matrices, scatter plots

### 6. **Target Variable Analysis**
- **What it is:** Understanding your prediction target
- **What you learn:** Class distribution, imbalance, patterns
- **How to do it:** Value counts, pie charts, bar plots

### 7. **Time Series Analysis**
- **What it is:** Understanding temporal patterns
- **What you learn:** Trends, seasonality, anomalies over time
- **How to do it:** Time plots, seasonal decomposition

### 8. **Categorical Analysis**
- **What it is:** Understanding categorical variables
- **What you learn:** Categories, frequencies, relationships
- **How to do it:** Value counts, cross-tabulations, bar plots

## 1.5 EDA Best Practices

### Do's ‚úÖ
- **Start simple** - Begin with basic overview, then dive deeper
- **Visualize everything** - Charts often reveal what numbers don't
- **Document findings** - Keep notes of what you discover
- **Ask questions** - What patterns do you see? What's surprising?
- **Validate assumptions** - Check if your initial thoughts are correct

### Don'ts ‚ùå
- **Don't skip missing values** - They can indicate important patterns
- **Don't ignore outliers** - They might be the most interesting cases
- **Don't assume normality** - Many security datasets are skewed
- **Don't rush** - Take time to really understand your data
- **Don't work in isolation** - Share findings with your team

---

## 1.6 Common EDA Challenges in Security

### 1. **Class Imbalance**
- **Problem:** Most security events are normal, few are attacks
- **Solution:** Use appropriate metrics and sampling techniques

### 2. **High Dimensionality**
- **Problem:** Many features, making analysis complex
- **Solution:** Feature selection and dimensionality reduction

### 3. **Temporal Dependencies**
- **Problem:** Security events often depend on time
- **Solution:** Time series analysis and temporal features

### 4. **Concept Drift**
- **Problem:** Attack patterns change over time
- **Solution:** Regular monitoring and model updates

### 5. **Data Quality Issues**
- **Problem:** Missing values, errors, inconsistencies
- **Solution:** Data cleaning and validation procedures

## 2.3 Data Quality Assessment

Now let's check for data quality issues that might affect our analysis.

### Task 2: Missing Values Analysis
**Your turn!** Analyze missing values in your dataset.


````python
print("üîç MISSING VALUES ANALYSIS:")
print("=" * 50)

# 1. Check for missing values
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

# 2. Create a summary table
missing_df = pd.DataFrame({
    'Column': missing_values.index,
    'Missing Count': missing_values.values,
    'Missing Percentage': missing_percentage.values
})

# 3. Filter to show only columns with missing values
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)

if len(missing_df) > 0:
    print("‚ö†Ô∏è Missing values found:")
    display(missing_df)
else:
    print("‚úÖ No missing values found!")

# 4. Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"\nüîÑ Duplicate rows: {duplicates}")

# 5. Check for infinite values in numerical columns
infinite_values = np.isinf(df.select_dtypes(include=[np.number])).sum().sum()
print(f"‚ôæÔ∏è Infinite values: {infinite_values}")

# 6. Check for outliers in numerical columns
print("\nüîç OUTLIER ANALYSIS:")
print("-" * 30)
continuous_columns = ['web_traffic', 'age_of_domain', 'Page_Rank', 'URLURL_Length']

for col in continuous_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_percentage = (len(outliers) / len(df)) * 100

    print(f"{col}: {len(outliers)} outliers ({outlier_percentage:.1f}%)")
````

## 2.4 Target Variable Analysis

Now let's analyze our target variable (attack_type) to understand the distribution of security events.

### Task 3: Target Variable Distribution
**Your turn!** Analyze the distribution of attack types in your dataset.


````python
print("üéØ TARGET VARIABLE ANALYSIS:")
print("=" * 50)

target_column = 'Result'

label_map = {-1: "Legitimate", 1: "Phishing"}
df['Result_label'] = df[target_column].map(label_map)

# 1. Count and percentage of each class
target_counts = df['Result_label'].value_counts()
target_percentages = df['Result_label'].value_counts(normalize=True) * 100

# 2. Create analysis table
target_analysis = pd.DataFrame({
    'Count': target_counts,
    'Percentage': target_percentages
})

display(target_analysis)

# 3. Check for class imbalance
max_percentage = target_percentages.max()
min_percentage = target_percentages.min()
imbalance_ratio = max_percentage / min_percentage

print(f"\n‚öñÔ∏è Class Imbalance Analysis:")
print(f"Most frequent class: {max_percentage:.1f}%")
print(f"Least frequent class: {min_percentage:.1f}%")
print(f"Imbalance ratio: {imbalance_ratio:.1f}:1")

if imbalance_ratio > 10:
    print("‚ö†Ô∏è High class imbalance detected!")
elif imbalance_ratio > 5:
    print("‚ö†Ô∏è Moderate class imbalance detected!")
else:
    print("‚úÖ Balanced dataset!")

# 4. Create visualizations
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Bar plot
target_counts.plot(kind='bar', ax=ax1, color='skyblue', edgecolor='black')
ax1.set_title('Phishing vs Legitimate - Count', fontsize=14, fontweight='bold')
ax1.set_xlabel('Attack Type', fontsize=12)
ax1.set_ylabel('Count', fontsize=12)
ax1.tick_params(axis='x', rotation=45)

# Pie chart
target_percentages.plot(kind='pie', ax=ax2, autopct='%1.1f%%', startangle=90)
ax2.set_title('Phishing vs Legitimate - Percentage', fontsize=14, fontweight='bold')
ax2.set_ylabel('')  # Remove y-label for pie chart

plt.tight_layout()
plt.show()
````

## 2.5 Feature Analysis

Now let's explore the individual features in our dataset to understand their characteristics.

### Task 4: Feature Characteristics
**Your turn!** Analyze the features in your dataset.


````python
print("üîç FEATURE ANALYSIS:")
print("=" * 50)

# 1. Analyze categorical features
print("üìä CATEGORICAL FEATURES:")
print("-" * 30)
categorical_columns = df.select_dtypes(include=['object']).columns

for col in categorical_columns:
    if col != target_column:
        print(f"\nüìä {col.upper()}:")
        print(f"Unique values: {df[col].nunique()}")
        print(f"Most frequent: {df[col].mode().iloc[0]} ({df[col].value_counts().iloc[0]} times)")
        print("Value counts:")
        print(df[col].value_counts(normalize=True).head(10))  # Show distribution %

# 2. Analyze numerical features
print("\nüî¢ NUMERICAL FEATURES:")
print("-" * 30)
numerical_columns = df.select_dtypes(include=[np.number]).columns

summary = df[numerical_columns].describe().T[['mean', 'std', 'min', 'max']]
display(summary)

# Detailed analysis only for continuous columns
continuous_columns = ['web_traffic', 'age_of_domain', 'Page_Rank', 'URLURL_Length']
for col in continuous_columns:
    if col in df.columns:
        print(f"\nüìä {col.upper()}:")
        print(f"Mean: {df[col].mean():.2f}")
        print(f"Median: {df[col].median():.2f}")
        print(f"Std: {df[col].std():.2f}")
        print(f"Min: {df[col].min():.2f}")
        print(f"Max: {df[col].max():.2f}")

# 3. Create distribution plots for numerical features
n_cols = 2
n_rows = (len(continuous_columns) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 5*n_rows))
axes = axes.flatten()

for i, col in enumerate(continuous_columns):
    if col in df.columns:
        axes[i].hist(df[col], bins=30, alpha=0.7, color='skyblue', edgecolor='black')
        axes[i].set_title(f'Distribution of {col}', fontweight='bold')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Frequency')
        axes[i].grid(True, alpha=0.3)

# Hide empty subplots
for i in range(len(numerical_columns), len(axes)):
    axes[i].set_visible(False)

plt.tight_layout()
plt.show()
````
